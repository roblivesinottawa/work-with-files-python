```python
from pathlib import Path
```

```python
entries = Path('../')
entries = Path.home()
```

```python
def print_directory_contents():
	entries = Path.cwd()
```

```python
for entry in entries.iterdir():
		print(entry.name)
		print(entry.parent)
		print(entry.parent.parent)
		print(entry.stem)
		print(entry.suffix)
		print(entry.stat())
		print()
```

```python
if __name__ == "__main__":
	print_directory_contents()
```
